## Load the data

### Subtask:
Loading the "/Student_performance_data _.csv" file into a pandas DataFrame.


```python
import pandas as pd

df = pd.read_csv("/Student_performance_data _.csv")
```

```python
df.head()
```

## Explore and preprocess the data



```python
print(df.info())
print(df.isnull().sum())
df.head()
```


Based on the output, there are no missing values in the dataset. The columns 'Gender', 'Ethnicity', 'ParentalEducation', 'Tutoring', 'ParentalSupport', 'Extracurricular', 'Sports', 'Music', and 'Volunteering' appear to be categorical. I will one-hot encode these categorical features, separate the features (X) and the target variable (y), and then split the data into training and testing sets.



```python
from sklearn.model_selection import train_test_split
import numpy as np

categorical_features = ['Gender', 'Ethnicity', 'ParentalEducation', 'Tutoring', 'ParentalSupport', 'Extracurricular', 'Sports', 'Music', 'Volunteering']
df_encoded = pd.get_dummies(df, columns=categorical_features, drop_first=True)

X = df_encoded.drop(['StudentID', 'GPA', 'GradeClass'], axis=1)
y_regression = df_encoded['GPA']
y_classification = df_encoded['GradeClass']

X_train, X_test, y_regression_train, y_regression_test = train_test_split(X, y_regression, test_size=0.2, random_state=42)
X_train, X_test, y_classification_train, y_classification_test = train_test_split(X, y_classification, test_size=0.2, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_regression_train:", y_regression_train.shape)
print("Shape of y_regression_test:", y_regression_test.shape)
print("Shape of y_classification_train:", y_classification_train.shape)
print("Shape of y_classification_test:", y_classification_test.shape)
```

## Introduce linear regression

### Subtask:
Explain the concept of Linear Regression and its applications.


**Reasoning**:
Provide an explanation of Linear Regression, including its concept, mathematical forms, and applications, as requested in the subtask instructions.



```python
print("Linear Regression Explained:\n")

print("1. Concept:")
print("Linear Regression is a statistical method used to model the linear relationship between a dependent variable (the variable we want to predict) and one or more independent variables (the variables used for prediction). The goal is to find the best-fitting straight line (or hyperplane in higher dimensions) that describes how the dependent variable changes as the independent variables change.")
print("\n2. Mathematical Forms:")
print("   - Simple Linear Regression (one independent variable):")
print("     The equation is typically represented as: y = mx + b")
print("     - y: The dependent variable.")
print("     - m: The slope of the line, representing the change in y for a one-unit change in x.")
print("     - x: The independent variable.")
print("     - b: The y-intercept, representing the value of y when x is 0.")
print("\n   - Multiple Linear Regression (multiple independent variables):")
print("     The equation is an extension of simple linear regression:")
print("     y = b0 + b1*x1 + b2*x2 + ... + bn*xn")
print("     - y: The dependent variable.")
print("     - b0: The y-intercept (or bias term), representing the value of y when all independent variables are 0.")
print("     - b1, b2, ..., bn: The coefficients (or weights) for each independent variable (x1, x2, ..., xn). Each coefficient represents the change in y for a one-unit change in the corresponding independent variable, holding other variables constant.")
print("     - x1, x2, ..., xn: The independent variables.")
print("\n3. Applications:")
print("Linear Regression is widely used in various fields, including:")
print("   - Economics: Predicting economic growth, inflation rates, or unemployment based on various indicators.")
print("   - Finance: Forecasting stock prices, analyzing the relationship between risk and return.")
print("   - Marketing: Predicting sales based on advertising spending, analyzing the impact of promotions.")
print("   - Healthcare: Predicting patient outcomes based on various factors like age, weight, and treatment.")
print("   - Real Estate: Predicting house prices based on features like size, location, and number of bedrooms.")
print("   - Education: Analyzing the relationship between study time, absences, and student performance (like GPA).")
```

## Implement and train linear regression

### Subtask:
Building  and training a Linear Regression model on the data.


```python
from sklearn.linear_model import LinearRegression

# Instantiate a Linear Regression object
linear_regression_model = LinearRegression()

# Fit the model to the training data
linear_regression_model.fit(X_train, y_regression_train)
```

## Evaluate linear regression




Importing the necessary regression evaluation metrics, make predictions using the trained linear regression model, calculate MAE, MSE, and R-squared, and print the results.



```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_regression_pred = linear_regression_model.predict(X_test)

mae = mean_absolute_error(y_regression_test, y_regression_pred)
mse = mean_squared_error(y_regression_test, y_regression_pred)
r2 = r2_score(y_regression_test, y_regression_pred)

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"R-squared (R2): {r2:.4f}")
```

## Introduce logistic regression

### Subtask:
Explaing the concept of Logistic Regression and its applications, particularly for classification tasks.


**Reasoning**:
Print the explanation of Logistic Regression, its mathematical form, and applications as requested in the instructions.



```python
print("Logistic Regression Explained:\n")

print("1. Concept:")
print("Logistic Regression is a statistical model used for binary classification problems, meaning it predicts the probability that an observation belongs to one of two classes. Unlike Linear Regression which predicts a continuous outcome, Logistic Regression predicts the probability of a categorical outcome. It models the relationship between the independent variables and the probability of the dependent variable belonging to a particular class.")
print("\n2. Mathematical Form (Sigmoid Function):")
print("The core of Logistic Regression is the sigmoid function (also known as the logistic function). This function maps any real-valued number to a value between 0 and 1, which can be interpreted as a probability.")
print("The sigmoid function is defined as: (z) = 1 / (1 + e^(-z))")
print("Where:")
print("   - (z): The output of the sigmoid function, representing the predicted probability.")
print("   - e: The base of the natural logarithm (approximately 2.71828).")
print("   - z: A linear combination of the independent variables and their coefficients, similar to the linear equation in Linear Regression: z = b0 + b1*x1 + b2*x2 + ... + bn*xn")
print("The output of the sigmoid function, (z), is then used to classify the observation. If (z) is greater than a certain threshold (commonly 0.5), the observation is assigned to one class, otherwise to the other class.")
print("\n3. Applications:")
print("Logistic Regression is widely used in various classification tasks, including:")
print("   - Healthcare: Predicting the probability of a disease based on symptoms and patient history.")
print("   - Finance: Predicting the likelihood of loan default based on credit score and financial history.")
print("   - Marketing: Predicting whether a customer will click on an advertisement or purchase a product.")
print("   - Spam Detection: Classifying emails as spam or not spam based on their content.")
print("   - Image Recognition: Classifying images into different categories (though deep learning models are more common for complex image tasks).")
print("   - Education: Predicting whether a student will pass or fail a course based on factors like study time and attendance.")
```

## Implement and train logistic regression

### Subtask:
Building and training a Logistic Regression model on the data.



Build and train a Logistic Regression model on the training data as instructed.



```python
from sklearn.linear_model import LogisticRegression

# Instantiate a Logistic Regression object
logistic_regression_model = LogisticRegression(max_iter=1000)

# Fit the Logistic Regression model to the training data
logistic_regression_model.fit(X_train, y_classification_train)
```

## Evaluate logistic regression

### Subtask:



Calculate and print the accuracy, F1-score, and confusion matrix for the logistic regression model's predictions on the test set.



```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

# Make predictions on the test set
y_classification_pred = logistic_regression_model.predict(X_test)

# Calculate and print Accuracy
accuracy = accuracy_score(y_classification_test, y_classification_pred)
print(f"Accuracy: {accuracy:.4f}")

# Calculate and print F1-score (weighted for multi-class)
f1 = f1_score(y_classification_test, y_classification_pred, average='weighted')
print(f"F1-score (weighted): {f1:.4f}")

# Calculate and print Confusion Matrix
conf_matrix = confusion_matrix(y_classification_test, y_classification_pred)
print("\nConfusion Matrix:")
print(conf_matrix)
```
